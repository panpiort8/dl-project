```python
# https://huggingface.co/models?pipeline_tag=text-generation&p=1
    
MODEL_SIZE = {
    "gpt2": ...,
    "gpt2-medium": 2_046_043,
    'tiny-gpt2': 3_911,
}
```

```python
import wandb
import pandas as pd


api = wandb.Api()

# Change oreilly-class/cifar to <entity/project-name>
runs = api.runs("dl-project2/universal-computation-engine")

results = []
for run in runs:
    if 'exp-model-size' in run.tags:
#         print(run.name, run.tags)
#     print(run.name, run.tags)
        state = run.name, run.config['task']+"-"+str(run.config['n']), run.config['model_name'], run.summary.get('Final Accuracy')
        #print(state)
        results.append(state)

df = pd.DataFrame(results, columns=['name', 'task', 'model_name', 'final_accuracy'])
df
```

```python
df.groupby(['task', 'model_name']) \
    .agg({'name': 'size', 'final_accuracy': 'mean'}) \
    .rename(columns={'name': 'count', 'final_accuracy': 'mean_accuracy'})
```
